# Step 1 - Data Ingestion

```python
url = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/refs/heads/main/PowerPlant.csv"
print(url)
```

```python
import pandas as pd
df = pd.read_csv(url)
df.head()
```

## Step 2 - Basic Data quality checks

```python
df.info()
```

1. AT - Atmospheric Temperture in degree celsius
2. V - Vacuum Pressure bars
3. AP - Atomospheric Pressure bars
4. RH - Relative Humidity %
5. PE (Target) - Power Output in Mega Watt

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(keep="first").reset_index(drop=True)
df.shape
```

```python
df
```

### Step 3 - Seperate X and Y

```python
X = df.drop(columns = ["PE"])
Y = df[["PE"]]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4 - Apply Preprocessing on X

```python
X.dtypes
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
).set_output(transform="pandas")
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre
```

### Step 5 - Train Test split

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtrain.shape[1]
```

### Step 6- Build the neural network

image.png

```python
from keras.models import Sequential
from keras.layers import Input, Dense
```

```python
model = Sequential(
    [
        Input(shape=(xtrain.shape[1],)),
        Dense(8, activation="relu"),
        Dense(4, activation="relu"),
        Dense(1, activation="linear")
    ]
)
```

```python
model.summary()
```

```python
model.compile(loss="mse", optimizer="adam", metrics=["mae"])
```

```python
hist = model.fit(xtrain, ytrain, epochs=60, validation_split=0.2)
```

### Check learning curve

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(hist.history["loss"], label="Train Loss")
plt.plot(hist.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.title("Learning Curve")
plt.legend()
plt.show()
```

### Model evaluation

```python
model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```

```python
from sklearn.metrics import (
    root_mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
)
```

```python
def evaluate_model(model, x, y):
  ypred = model.predict(x)
  rmse = root_mean_squared_error(y, ypred)
  mae = mean_absolute_error(y, ypred)
  mape = mean_absolute_percentage_error(y, ypred)
  r2 = r2_score(y, ypred)
  print(f"RMSE : {rmse:.2f}")
  print(f"MAE : {mae:.2f}")
  print(f"MAPE : {mape:.2%}")
  print(f"R2 : {r2:.2%}")
```

```python
print("Train Results :")
evaluate_model(model, xtrain, ytrain)
```

```python
print("Test Results :")
evaluate_model(model, xtest, ytest)
```
